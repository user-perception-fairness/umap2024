```python
respostas = [

" Não gostei dos filmes recomendados.",
" Caso não levasse em consideração o desejo e preferências da pessoa questionada.",
" Quando não levassem em consideração os gêneros que eu mais avaliei",
" Justas!",
" Acho que poderia ter mais uma filtragem, menos opções no final, eu sou indecisa ter dez filmes ainda é difícil pra mim escolher kkk",
" Nenhum, pois achei todas juntas e diferentes",
" Achei injusta por ter poucas opções de filmes para avaliar no início, acabou tento poucos filmes que realmente gosto.",
" Não consegui compreender a pergunta!",
" Teve alguns que não assisti",
" Não considerei as opções de escolha desse sistema injusta, inclusive uma boa parte das recomendações eu já tinha assistido, achei bem eficaz.",
" Justa.Em um momento de indecisão",
" São justas quando escolhem filmes parecidos dos quais eu assisto e são injustos por mostrarem apenas filmes mais famosos, sendo que existem filmes latinos ou turcos, por exemplo, com temáticas parecidas, só que pouco conhecidos.",
" Formas justas é entregar tanto filmes de grandes orçamento quanto de pouco orçamento,mesmo sendo talvez menos conhecido.",
" Justas quando recomenda filmes e gêneros de filmes que gosto de assistir. Injusta quando recomenda filmes que não tem a menor possibilidade de assistir.",
" Para o meu perfil os filmes de terro e suspense.",
"Injusta se tiver empresa interessada em determinado filme.",
" Achei justas as escolhas de filmes recomendada, pois levou em consideração minhas preferências.",
" Com base na avaliação prévia desta pesquisa, achei justa as recomendações feitas pela IA.",
"  Consideraria injusta se fugisse totalmente das informações que IA coletou minhas, e não tivesse nada a ver com o que consumo. Em geral, se for sugestãos que acrescente ainda mais naquilo que já gosto de consumir, facilitaria demais e seria ótimo.",
" As recomendações foram justa, apesar de 90% dos filmes serem filmes q eu já assisti, seriam filmes q eu escolheria caso estivesse procurando aleatoriamente. Mas a única coisa q pra mim foi fora do padrão e q a maioria dos filmes recomendados, forma filmes bem conhecidos.",
" Acho que são injustas quando as recomendações não são boas ou não sou o perfil da pessoa",
" Seria injusto que as recomendações só ocorram para filmes de grandes produtoras ou muito hypados",
" Não achei injustas, só que das 10 recomendações, 8 eu já havia assistido.",
" Não acredito que o algoritmo seja injusto.",
" Eu achei as recomendações bem justas! E amaria ter um site ou app desse para me ajudar quando quisesse escolher um filme! Achei bem prestativo.",
" Quase Todos os filmes sugeridos eu já assisti então as recomendações foram justas sim",
" Injusto recomendar filmes que já tinha marcado que assisti",
" .",
" Achei que as opções foram bem variadas, em cima dos meus filmes antes avaliados, o que torna as recomendações justa. Seria injusto se fosse selecionando como recomendações apenas uma temática de filme.",
" Achei as escolhas justas, estilos de filmes que eu gosto muito de assistir. Não seria justa se aparecesse opções de filmes que não se encaixam ao meu gosto. Inclusive as recomendações foram muito boas, tanto que somente um dos recomendados eu não havia assistido ainda.",
" seriam injustas caso as recomendações de filmes não combinassem com os gêneros que avaliei.",
" Provavelmente se eu escolhesse mais opções de animações ou romances e recebesse somente recomendações de filmes de ação ou drama, provavelmente eu acharia injusto a seleção de filmes.",
" Só avaliei uma animação na primeira fase da pesquisa e recomendaram várias. De resto achei as recomendações bem de acordo com os filmes que avaliei melhor (suspenses e dramas).",
" Recomendar todos os filmes de uma franquia de uma vez não me pareceu efetivo, pois, se já assisti um e gostei provavelmente verei os outros. Isso Tira a possibilidade de recomendar outros filmes do mesmo gênero que talvez eu não conheça.",
" Eu consideraria uma recomendação justa quando se leva em consideração não apenas a preferência pessoal, mas também estado de ânimo da pessoa no momento da procura, pois por mais que eu goste de filme de ação e aventura, as vezes quero assistir uma comédia no estilo minha mae é uma peça. O psicológico da pessoa sempre vai tender de um lado ou de outro quando se está procurando algo pra comer por quê o mesmo não vale pra quando se quer assistir algo?",
" Torna-se injusto quando não leva em consideração o gosto do telespectador nas primeiras perguntas",
" Algumas recomendações realmente são compatíveis com minhas preferências, já outras nao",
" Injustas se fossem mais filmes indicados de gêneros que não marquei.",
" Em casos que eu não tenha marcado a opção de algum filme e se aparecesse algo que não gosto.",
" Injustas se eu avaliasse um filme de forma negativa e, viesse a ser recomendado para mim filmes sinônimos a ele.",
" As recomendações seriam injustas se não considerassem o meu gosto pessoal. Como foram geradas com base nos filmes que já assisti e avaliei positivamente, foram indicações justas.",
" Foram justas, levando em consideração que a IA me recomendou filmes da mesma franquia, além de e oferecer outras versões antigas do mesmo filme e me indicou um filme que sempre julguei ser clichê, mas ao ler a sinopse mudei de ideia rapidamente. Hoje vou encerrar o meu dia assistindo O diário de uma paixão",
"É incrível como a IA pode tomar decisões e me recomenar filmes que tenha um dos períodos que mais gosto de estudar, II Grande Guerra.",
" Não entendi a pergunta, confusa",
" Injustas se recomendar apenas um gênero de filme . E justa se houver uma ampla diversidade de opções.",
" Justa pois é o nosso gosto q é refletido mas em vários casos esseas inteligências excluam as menores produções",
" Seriam injustas se não levassem em conta minhas preferências, mas eu gostei bastante das recomendações que me deram.",
" Os filmes selecionados foram adequados ao meu gosto. A maioria dels eu ja havia visto.",
" Eu não assisti a maioria dos filmes recomendados.",
" Nao tenho uma opinião formada da sobre isso porque nunca usei, essa foi minha primeira experiencia e gostei bastante, as escolhas pra mim foi bem nos filmes que eu gosto por causa dos meu filhos pequenos, apesar de ja ter assistidos todos que o sistema me apresentou . Mas gostei da experiência.",
" Eu acho injusta porque se baseia em único seguimento de gênero. Não ter gostado de um filme de suspense específico, não significa que eu não goste de suspense. E eu ter gostado de um filme de comédia, não significa que eu gosto de todos os filmes de comédia. As opções de escolha no início eram bem limitaras e eu não conhecia vários.",
" Injusta",
" Estou satisfeita pelos filmes apresentados, me deixando, inclusive, com muita vontade de assistir os filmes sugeridos.",
"  Injustas quando me recomendou muitos filmes da mesma série de filmes, no caso do velozes e furiosos, por presumir que por ter assistido um filme da franquia gostaria de assistir outros filmes da mesma franquia, contudo me recomendou os filmes anteriores do que eu classifiquei. Justo quando me recomendou filmes do gênero que eu demonstrei interesse, mas filmes com baixo investimento, então pouco conhecidos mas dentro do estilo indicado.",
" Acho injusta, em questão de não ter tantas opções, pois eu posso gostar de um gênero de filme mas não gostar do que apareceu pra colocar nota.",
" Como um todo achei as recomendações boas, tanto é que já havia assistido algumas das recomendações",
" Achei justas as recomendações, afinal estavam de acordo com lá filmes que avaliei no comeco",
" Eu penso que o algoritmo entende o gosto das pessoas e pode sim funcionar da melhor forma possível porque ele me deu boas opções. Porém, todas as opções que ele recomendou foram referentes a um único gênero de filmes, mesmo eu tendo escolhido outros também. Sua escolha foi baseada na maior porcentagem do mesmo gênero que eu classifiquei. Eu não pesquisei por nenhum, e sim, procurei os 7 primeiros que estavam no catálogo que eu já tivesse assistido. Senti que o algoritmo podia levar também em consideração os outros gêneros ou até mesmo o tipo de assunto abordado em cada filme votado.",
" Baseado nas minhas pesquisas",
" Consideraria injustas se os filmes recomendados não se encaixassem nas minhas preferências",
" Acredito que seria uma forma justa de proceder com algoritmos se eles fossem solicitados para atuação, porém, quando consentimos e aceitamos os termos de uso de vários sites e aplicativos damos a liberdade de esses algoritmos trabalharem por si só. Então, não acho correto a mídia afirmas que esses algoritmos são injustos quanto ao aparecimento de recomendações ou propagandas relacionadas as mossas pesquisas e interesses.",
" O algoritmo me recomendou filmes conhecidos e poucos deles eram atuais, é injusto não recomendar filmes menos conhecidos e mais atuais.",
" Justa quando você seleciona os gêneros que quer assistir e ele tem entrega o que você pediu, injusta quando você perde um certo tempo marcado e no final ele tem entrega nada com nada. Não é que a gente queria que o sistema de a resposta perfeita mas sim a que mais se aproxima do que a gente quer.",
" Acho injusta a recomendação de filmes que já estão em alta ou são amplamente divulgados pela mídia.",
" Não sei",
" Não acho que sejam justas ou injustas.",
" Acredito que como se trata de filmes relacionados a gostos e gêneros não teria um grande problema pois as únicas situações injustas com inteligência artificial que eu me deparei na internet foram sobre questões raciais onde o banco de dados só estava relacionado a mais pessoas brancas do que uma melanina mais escura. Ou seja tudo depende de como essas inteligências foram programadas para serem usadas. A ideia de ter uma inteligência que recomende filmes baseado no seus gostos é bem bacana, muitas pessoas não tem ideia do querem assistir ou casos que querem conhecer filmes novos mas não sabem quais opções vão gostar e acabam desistindo, portanto é uma excelente alternativa.",
" Não gostei muito das opções",
" Não entendi",
" Filmes e sugestões que fogem dos meus interesses pessoais..",
" Injustas seriam somente filmes famosos por exemplo, gosto muito de terror e grande parte das vezes os algoritmos ja existentes tem tendencia a mostrar filmes mais famosos e deixa de lado alguns filmes menos famosos que são muito bons, principalmente os que nao sao americanos ou ingleses",
" Concordo com o gênero, porém os filmes são bons não",
" As recomendações seriam injustas caso dependessem do orçamento do filme para serem feitas, ou seja, quanto maior o orçamento maior a recomendação.",
" Quando ele gera recomendações que são totalmente difentes do meu gosto pessoal.",
" Não sei responder a tal pergunta.",
" Apesar de assistir filmes de terror, não são o meu gênero preferido, então teria que ter mais opções de gêneros diferentes",
" Consideraria injustas se me indicassem filmes totalmente diferentes do que avaliei ou que me indicasse filme semelhantes ao que avaliei mal",
" Considero injusta no caso do filme Crepúsculo onde avaliei todos os outros filmes da franquia e me foi recomendado o único que não foi avaliado anteriormente, mas mesmo assim não considero como uma recomendação ruim.",
" Me atendeu em todos casos, achei sensacional e acho que beneficiária muito.",
" Muitos dos filmes recomendados eram sequencia dos filmes selecionados ou do mesmo estúdio. Gostaria de ter visto mais filmes com a mesma pegada mas nao necessariamente relacionado.",
" Não gosto de filmes de terror extremo. E nesse caso seria injusto quando marquei muitos filmes de desenho. Mas em geral, as recomendações foram justas e boas.",
" Um dos filmes que avaliei foi Era do gelo 3, e o algoritmo me indicou Era do Gelo 1, não é totalmente injusto, mas entende-se ser uma sequência de filmes, se eu assistir o 3, a possibilidade de já ter assistimos o 1 e 2 é muito grande.",
"Recomendações justas na minha opinião seria:por razão de ser o mesmo diretor, ator-atriz principal,categoria.Recomendações injustas seria se forem em base de popularidade em redes sociais em geral, arrecadação em bilheterias.",
" Injustas = quando só é recomendado filmes recém lançados",
" Se apenas um gênero fosse recomendado não seria justo. Acredito que abrir o leque de opções é uma boa ideia, tanto para mim quanto para o sistema que me recomenda",
" Quando o sistema for totalmente imparcial aos meus gostos. Acredito que isso seria um problema desde que meu gosto seja totalmente distorcido e acabo tendo que optar por filmes que não assistiria.",
" Acho que as recomendações podem ser injustas quando feitas com base no orçamento da produção ou em que tenha um elenco famoso.",
" A maior parte dos filmes recomendados não são tão conhecidos, por isso, não foi possível avaliar tão bem se a recomendação foi boa ou não. Ademais, algumas perguntas são difíceis de responder tendo em vista que os filmes não foram vistos. Mas achei a proposta muito boa e utilizaria novamente!",
" Na minha opinião os filhos são justa",
" Seriam justas desde que não se baseassem em critérios desonestos",
" Consideraria injusta se não fosse abrangente aquilo que eu procuro, caso contrário pra mim seria justo.",
" Não sei como opinar",
" Ao meu ver o grande problema de se basear em na inteligência artificial o tempo inteiro é que ele vai acabar limitando as suas opções e ficando difícil de conhecer novos gêneros e filmes mas ao mesmo tempo não acho isso injusto pois somente alguém muito ligado a essa indústria iria se sentir prejudicado, a grande maioria da população não se importa com isso, só quer chegar em casa depois de um dia longo de trabalho e ver um filme, se a própria IA facilita essa escolha ao meu ver é um grande benefício pois os mesmos não querem perder muito tempo para escolher um filme.",
" .",
" Nos casos onde seria recomendado filmes repetidos e que não façam parte do meu perfil e preferência",
" Já havia assistido",
" Se recomendassem apenas um tipo específico de filmes poderia ser problemático por outro lado seria bom para quem se prende na mesma coisa o tempo todo.",
" Muitas vezes esses algoritmos podem nos influenciar sem que percebemos, mas algumas vezes podem trazer alguns benefícios",
" Essa seleção é injusta,pois o algoritmo acaba excluindo opções interessantes e privando a possibilidade de conhecer algo além das preferências.",
" Nenhuma, o sistema indicou bons filmes, que com certeza eu assistiria",
" Uma recomendação injusta seria se os filmes indicados não tivessem mesmo gênero, por exemplo, dos filmes que havia selecionado anteriormente. Achei a seleção muito justa, pois manteve meus gostos - tanto que algumas recomendações eu já assisti.",
" Acho injusto filmes de baixo orçamento não serem indicados, parece que só filmes indicados ao oscar são merecedores de ser vistos...",
" Quando recomenda sequências. Por exemplo kill bill 2 ou sobrenatural 2. Isso ja e algo que provavelmente eu vou procurar para assistir. É melhor me recoemndar coisas q eu nem sei e conheço e provavelmente vou gostar...como outros filmes do mesmo diretor",
" Caso tivessem me indicado filmes baseado apenas em um estilo selecionado, como por exemplo se tivessem indicado somente animação.. é importante variar assim como foi feito aqui.",
" Acredito que seria justo mostrar filmes pouco conhecidos que se baseiam nos gostos que a pessoa falou, assim acaba dando mais visibilidade a filmes bons com uma história ótima mas que não tem o conhecimento que mereciam.",
" Particularmente, odeio finais tristes. Então para me fazer uma excelente sugestão de filme, além do enredo, esse fator desse considerado.",
" Se a seleção é em base nossas preferências não vejo motivos para ser injusto não viabilizar os demais. Hoje, várias plataformas exibem diversos títulos e se tornaria uma experiência mais agradável com indicações de filmes, baseado com o que nos identificamos. Injusto é assistir um filme de 1h40, se conseguir concluir, e não gostar, ser uma perca de tempo que poderia ter sido otimizado, com o auxílio da I.A.",
" Consideraria injusta se a recomendação só fosse de filmes em alta, de alto orçamento. Seria justo a recomendação de filmes menos conhecidos, mesmo tendo um baixo orçamento, sendo uma indicação por boas atuações.",
" Acho q um aplicativo de recomendações de filmes pelos seus gostos já pré selecionados não teria como dar recomendações injustas.",
" Foi recomendado as sequências de alguns filmes que eu já havia classificado, assim deixando de aparecer novos tipos, ao invés de se fixar em uma franquia, faltou explorar um pouco além destes",
" Injusto colocar um filme de terror ou suspense pata quem não gosta do gênero.",
" Alguns filmes de ação eu nunca vi ou não reconheci o título, alguns eu sei porque meu pai gosta desse gênero! Minhas preferências são comédia, romance, aventura e animações! Vi poucas opções nesses gêneros para escolher!",
"Nesse teste foi injusto pois em nenhum momento foi solicitado os gêneros que eu gosto.  Somente foi indicadonos filmes para escolha.",
" Recomendações de filmes injusto quando você consideram o filme ruim mas recomendam por popularidade",
" As recomendações foram óbvias porque seguiram o gênero geral dos filmes. Em filmes onde o gênero não é claramente definido, o algoritmo vai ter dificuldade em fazer uma indicação apropriada.",
" eu não gosto quando os sistemas de IA recomendam somente pelo hypedos filmes, mas quando eles aprendem os meus gostos eu fico bem satisfeita.",
" Acredito que as recomendações acabam sendo pouco injustas porque levam em consideração o que é popular. Todos os filmes que foram recomendados são populares dentre os que eu assisto, nenhum me surpreendeu. Mas, apesar disso, o algoritmo conseguiu entender os meus gostos e aí tenho menos chance de errar na escolha do filme.",
" Neste caso, a maioria dos filmes foram continuações dos filmes que eu havia escolhido, porém os que não eram continuações fazem total sentido pelo gênero que eu gosto e assisto.",
" Não sei responder a pergunta em questão.",
" Muitas das recomendações foram filmes que já havia assistido.",
"A partir do memento que você procura uma recomendação , obviamente se vai pelo gosto pessoal . Quando se seleciona opções que vai te levar por esse caminho próprio , seria uma recomendação justa , porém , quando você não leva em consideração outros fatores que não são selecionados , pode se dizer que o tiro no escuro seria uma recomendacão injusta.",
" Na realidade a primeira parte é que me pareceu injusta. Tive necessariamente que avaliar 7 filmes, sendo que três deles eu não assisti.",
" Consideraria injustas se me recomendasse filmes que eu não olharia, nem mesmo por recomendação.",
" Acho que nem sempre vai ser claro, as vezes assistimos algo apenas por assistir!",
" Poderia ser recomendação injusta em casos que as seleções fossem decididas a partir de temas de filmes. As vezes, gostamos da história, mas não do gênero e por aí vai.",
" Os filmes indicados tinha por base o gênero que mais assisti, no caso terror, thriller e animação, assim achei justa as indicações recebidas.",
" Acho super justo pois as vezes estamos indeciso ou não sabemos qual filme ver essa pode ser uma boa alternativa para gente tanto descobrir filmes novos ou rever eles",
" Creio que são justas as recomendações de filmes, até por que muitas das pessoas são bastante indecisas no que escolher para aproveitar seu tempo de descanso e tendo uma forma de como pode ajudar seria maravilhoso, tanto em forma de aplicativos como em forma de site.",
" As recomendações pra mim foram justas, é o geralmente assisto, a parte injusta o filme de 73 que nunca ouvi falar .",
" As recomendações foram justas.",
" No geral, a IA sugeriu muitos filmes que já tive curiosidade em assistir, então de certo modo, acertou minhas preferências, ou ao menos curiosidades. Acredito que sugerir filmes de baixo orçamento ou baixa popularidade é um ponto justo e positivo, fazendo com quem as pessoas tenham um maior conhecimento sobre filmes no gênero de seu interesse que não sejam necessariamente blockbuster",
" Levando em consideração que já tinha assistido a maioria dos filmes recomendados,enquanto poucos dos que foram apresentados primeiro, acredito que as recomendações foram muito boas e dentro do que realmente gosto de assistir.",
" A maioria eu não conheço. Somente os da Disney",
" Injustas caso não refletissem no meu gosto pessoal e histórico de filmes que marquei como assistidos e filmes com boas avaliações dadas por mim. Justas as recomendações de filmes que seguem esse critério.",
" No caso das recomendações serem totalmente diferentes dos filmes anteriores",
" São justas quando o algoritmo consegue realmente se basear em um seleção de filmes que assistimos.",
" Eu consideraria as recomendações de filmes injustas diante de uma troca de gênero muito brusca, como preferir filmes de comédia romântica e indicarem um thriller.  Considero justas as recomendações com similaridades às preferências.",
" Acredito que uma recomendação justa é aquela que condiz com o que a pessoa costuma assistir",
" .",
" Ele analisa a partir das informações que fornecemos. Não tem como apontar se está justo ou não uma seleção feita a partir de uma inteligência artificial.",
" Na verdade em nenhum caso, as recomendações estão de acordo com o que venho pesquisando e assistindo, e mesmo que seja recomendado algum filme fora do que geralmente assisto não vejo nada de ruim nisso, se eu ano quero assistir e só não assistir, as recomendações servem pata te dar um direcionamento melhor, não escolher por vc.",
" Justas quando são quando a maioria gostou e recomendou Injustas são falsas afirmações sobre os filmes serem bons ou não",
" Alguns filmes que aparecem nunca tinha ouvido falar",
" Acho que são justas quando recomendam filmes com o mesmo estilo de animação, gênero e ano de lançamento dos filmes que eu já gosto. Por exemplo, com base nos meus gostos seria injusto me recomendar um filme da Marvel ou de terror.",
" Injustas se me indicassem somente filmes de ação",
" Em aplicativos de streaming acho justo",
" …",
" Eu gostei muito pois todos os filmes que recomendou foram bons, porém a maioria é muito conhecido.",
" Justas quando o algoritmo já pegou que gosto bastante daquele gênero e me recomenda do.mesmo .",
 " Quando tem uma maior base de dados. Achei essa pouca. Poderia ter mais opções para ampliar o número de escolhas.",
" Não sei",
" Quando não está de acordo com meus gostos.",
" Justas quando indicasse filmes parecidos, ou do mesmo gênero aos que já assisti, injusto quando ocorresse o oposto disso. Por exemplo, não assisto filme de terror, não marcaria nenhum e nas recomendações aparecesse esse gênero.",
" No caso de as opções serem totalmente iguais ou totalmente longe do meu gosto",
" É justo quando o algoritmo considera a recomendação com base em um gênero, por exemplo: indicar outros filmes de romance por ter tido o expectador gostado de algum filme do gênero. Por outro lado, seria injusto indicar filmes com base num ator, simplesmente porque o expectador gostou de algum filme com ele. Basear o gosto simplesmente num ator, pode trazer diversos gêneros diferentes para as opções do expectador",
" Recomendações injustas: Duas das recomendações não assistiria de forma alguma.  Justas: Foram recomendados filmes de sequências de filmes que eu avaliei, o que seria útil caso eu ainda não tivesse visto os mesmos.",
" Nenhuma gostai das recomendações",
" Falta de opções de escolha",
" Na verdade, se for procurado por gênero, acredito que apareçam muitas sugestões que já foram vistas pelo telespectador. Mas pesquisando por mais de um gênero, acredito que as sugestões sejam de ótima ajuda!",
" Não consideraria injusto",
" Em casos onde as escolhas são guiadas apenas pelo estilo favorito selecionado anteriormente.",
" No caso de não ter nenhum filme que me interessa",
" Minhas preferências por filmes são animações, romance e comedia.Porém as indicações não tiveram muito a ver com essas preferências",
" Muito justas e baseadas nas minhas primeiras escolhas",
" Na tela inicial colocaria filmes mais conhecidos, para ajudar a seleção de filmes que gostaria de assistir.",
" Foram todas justas.",
" A partir do meu gosto, até então, encaixou perfeitamente e achei justo. Teve filme que nunca vi, mas já ouvi falar",
" Justas",
" Consideraria injustas se baseadas somente em poucas avaliações. Consideraria justas, se baseadas em mais algumas questões, como por exemplo uma maior quantidade de perguntas.",
" Justas. Só foram recomendados gêneros fora dos que eu avaliei. Tiveram recomendações de terror, por exemplo, sendo que grande parte dos filmes que avaliei foram animações infantis.",
" Filmes muito conhecidos ou sequências",
" Justa",
" Ao indicar o primeiro filme de uma franquia, quando já indiquei ter assistido o segundo, uma vez que obviamente já deveria ter assistido ao primeiro, não vejo necessidade de indicar o primeiro filme.",
" Os filmes que recomendaram, foram os mesmos marcados anteriormente. Tirando o que escolhi pra assistir, que foi o unico fora da lista.",
" Não vejo porque a indicação de filmes usando IA seria injusta",
" Justo quando apresenta recomendações diante dos interesses atuais do usuário, já que os algoritmos lêem o momento, mas injusto quando não abre um leque maior de opções, por exemplo na pesquisa, dentre os filmes apresentados na opção todos, que foi a que escolhi por gostar de filmes de categorias diversas, as recomendações foram boas e coerentes com as classificações que eu dei, mas a própria opção todos era limitada, não tinha por exemplo filmes de suspense e terror que dispertam também meu endereço.",
" Acho injusta a recomendação de filme, quando a indicação é de filme não combina com o gosto da pessoa, como os filmes que ela selecionou que gosta e assiste.",
" No meu caso eu selecionei filmes de comédia e romance e me indicou alguns filmes de terror",
" Eu não acho injusto o sistema de assistência virtual, pelo contrário, até porque sempre que precisei me foi recomendado filmes que geralmente não vemos publicações na mídia e que respeitam meu estilo preferido",
" As recomendações se tornam injustas no momento em que promovem demasiadamente um tipo de filme e excluem a oportunidade conhecer outros nomes que não são tão comerciais, ao mesmo tempo pode ser beneficial para uma economia de tempo na procura de filmes por excluir gêneros de filmes que não seriam escolhidos por você através da sua personalidade.",
" Quando me indicam filme de terror.",
" Não sei",
" Nesse caso as recomendações seguiram certinho de acordo com as minhas escolhas iniciais. Mas acredito que existam situações em que possam vir recomendações nem tão de acordo com o perfil do usuário, principalmente quando é um usuário que tenha um gosto mais eclético.",
    " Achei justa apenas de dois filmes dos quais foram selecionados para mim e achei injusto se basearam em tão poucas informações para fazer a seleção. Gosto para filmes deve-se ser feita perguntas mais específicas como, por exemplo, escolha três títulos que mais te surpreendeu ou que esperava não gostar, mas o filme superou a expectativa.  Acontece muito comigo, pois escolho um filme aleatório já com o pensamento de que não vou gostar, mas assisto mesmo assim, e o filme supera o meu pensamento anterior. Revejam essa pesquisa, pois não curti! Boa sorte com as melhorias e espero ter ajudado! Uma boa semana!",
" Injustas seriam as recomendações que não teriam nada a ver com os meus costumes de filmes vistos.",
" Seriam injustas caso só fossem recomendados filmes mais populares.",
" Não sei dizer",
" Gostei bastante das recomendações, hoje em dia buscamos sempre recomendações através da internet e o algoritmo tenta recomendar filmes, séries e livros de acordo com o que você mais interage e esse sistema de recomendação de filme também seguiu dessa forma ao recomendar os filmes mais semelhantes com os que avaliei bem. O que achei um pouco injusto foi o fato das recomendações ter só uma animação, sendo que os mais bem avaliado no início foram de animação.",
" Em filmes avaliados com poucas estrelas, se determinado gênero foi avaliado de forma ruim não é justo que haja uma sugestão dele ou de outro nesse mesmo gênero.",
" Em nenhum dos casos.",
" No caso em questão foram recomendações justas, pois ocorreu com base nas minhas avaliações em outros filmes. Recomendações injustas para mim são as que se baseia no que eu assisti independente deu ter gostado ou não, ficam recomendado diversos títulos semelhantes aos que assisti recentemente sem levar em consideração o que achei dos filmes.",
" Eu consideraria como justa tomando por base o gênero de filme que eu avaliei, pois indica que assisti, e a nota que eu dei, pois indica o quanto gostei ou não do filme. Assim, é justo que o algoritmo parta da análise dos meus gostos para me indicar filmes semelhantes. Injusto seria o algoritmo me indicar filmes de gêneros completamente diferentes daqueles que eu vi/analisei, pois ele não teria um ponto de partida de onde tirar as informações sobre o meu gosto pessoal. Poderia me sentir desconfortável, inclusive, pensando se o algoritmo coletou outros dados meus sem permissão.",
" Sobre os filmes iniciais, seria mais fácil encontrar uma forma de avaliação se eu tivesse assistido a maioria, mas a maioria dos filmes eu não conhecia. Então seria mais legal se na primeira avaliação vcs colocassem filmes famosos de cada genero, assim os filmes recomendados vão ser mais certeiros.",
" Seria injusto recomendar filmes de outros tipos que não foram muito votados Justo seria ter recomendações de filmes com base no q a pessoa já viu",
" De acordo com os filmes que avaliei anteriormente, achei bem condizente as opções de filmes recomendadas, tanto que a grande maioria eu já havia assistido por serem meu estilo.",
" Não achei nenhuma injusta. E achei justa as recomendações dadas pq foram de acordo com a minha seleção de filmes já assistidos.",
" No geral foi tudo bem só que tem alguns filmes na primeira página que são muito desconhecidos",
" Eu achei ótimo as recomendações, assistiria 90% do q me foi recomendado. Porém foram filmes que eu já assisti a maioria, eu assistiria novamente porque eu curto assistir filmes repetidos.",
" Caso fossem de apenas um estúdio ou diretor.",
" Quando a recomendação vem de gêneros que não foi escolhido",
" Seriam recomendações injustas, aquelas que não tivessem nada haver com o meu perfil.",
" Realmente teve varios filmes aos quais eu não assistiria.",
" O filme gravidade foi recomendo possivelmente por ser ficção científica. Mas ele possui personagens cientistas, se passa no espaço, mas eu não considero ele ficção científica, seria mais um Thriller. Ex Machina é um exemplo que considero ficção científica, por isso achei a recomendação injusta.",
" Achei injusta a indicação de filmes que eu já tinha avaliado.",
" Não acho que uma inteligência artificial escolhe as opções de forma justa ou injusta, ela apenas se baseia naquilo que você já assistiu ou procurou sobre, não é uma escplha injusta já que são recomendações com base naquilo que você já assistiu.",
" As recomendações foram justas e seguiram o que, de acordo com as minhas respostas preliminares (ou seja, de acordo com o meu “gosto”), faziam sentido para que eu escolhesse.",
" Considero injusta a recomendação de um gênero que eu não selecionei",
" É um robo , não sabe nada emocionalmente que me toca ou que gosto",
" Injustas para mim seria se as indicações fossem apenas filmes hypados. No entanto, no caso desse experimento teve um pouco de tudo.",
" Injustas se aparecessem filmes fora do padrão que eu escolhi, ex: uma biografia não se encaixaria em filmes que eu escolheria para assistir.",
" O meu caso, a maioria dos filmes que assusto são populares, portanto, para o algoritmo escolher uma recomendação seria difícil aparecer com algo que eu já nao tenha visto. Exemplo disso foram as recomendações da pergunta final em que eu tinha assistido a maioria e não tinha interesse nos de gênero que anteriormente escolhi apenas um",
" Injustas, onde por conta de um filme escolhido o algoritmo recomenda o mesmo estilo sempre, exemplo no streaming Netflix por conta de uma vez assistido um filme/série coreana ele só recomenda isso, mesmo assistindo outros estilos, não dando oportunidade para outros estilos.",
" eu consideraria quando o algoritmo me entregar filmes que não são baseados nos meus gostos pessoais",
" Seria injusto se selecionassem filmes com gêneros e ideias diferentes dos filmes escolhidos anteriormente",
" Só se for completamente contra as minhas opiniões anteriores.",
" Quando elas são de acordo com minhas pesquisas são justas e ampliam a visão apesar de serem do gênero que mais gosto",
"Injustas: recomendações baseadas apenas na coincidência de atores que atuaram em algum filme que o usuário gostou. Justas: histórias levemente semelhantes, mesmos pontos chaves (ex: filmes com foco na superação) em comum.",
" considerei todas as seleções justas e de acordo com aquilo que selecionei, o problema é que não consegui pesquisar filmes na primeira etapa",
" Considero injusto se o sistema der preferencia a grandes produções. Acho que é inclusive interessante que recomende mais filmes independentes",
" Gosto de sugestões por gênero, dão um leque maior para conhecer e escolher melhor",
" Indicar outro filme da franquia de filmes que eu selecionei como gostei",
" Não sei direito, mas acho que falta opções das quais realmente me faria querer assistir",
" Não sei opinar",
" Seria injusto considerar somente filmes famosos e com alto orçamentos. Mas analisando os filmes indicados o algoritmo propôs ótimos filmes.",
" De acordo com os filmes que avaliei, achei justa as recomendações que recebi. Apesar que as recomendações dos filmes fora um pouco previsíveis, pois entre elas tinha as sequências dos filmes que avaliei. Seria injustas as recomendações que não fossem de acordo com os gêneros que avaliei.",
" Justa quando recomenda algo parecido com o que já vi e avaliei bem. Injusto quando aparece algo do qual não avaliei bem e não é do gênero que selecionei como favorito.",
" Creio que os filmes de baixo orçamento tem poucas recomendações, e assim (quem gosta mesmo de cinema) acaba perdendo bons trabalhos que não são divulgados.",
" Eu definitivamente não sei",
" Apareceram sim alguns filmes q eu já assisti mas também apareceram alguns filmes nada haver então acho q funciona mais ou menos",
" Acho difícil a avaliação quando se tem poucas opções para definir os gêneros.",
"Geralmente leio que é injusto por ele te recomendar filmes parecidos com os que já assistiu, e isso te impede de ver coisas novas. Mas ao meu ver se tu possui uma preferência, iria querer ver mais desse tipo, então não me faz tanto sentido essa reclamação. Gostei das recomendações, e o fato de possuir filmes que tanto eu já vi, quanto já me foram recomendados por amigos em conversas onde falei justamente dos filmes que escolhi me fizeram levar ainda mais em conta as escolhas do algoritmo.",
"Injusta quando são recomendações de filmes antigos ou sempre as mesmas recomendações. Justa, quando apresenta uma variada maior de filmes, tanto antiga como novo,além de uma variedade maior.",
" Seria injusto se recomendasse filmes relacionados com outros filmes que eu tenha avaliado com nota baixa.",
" Eu não acho ruim, toda opção de sugestão é valida até que se prove o contrário. Eu gostei das opções, apesar de estar meio por fora do mundo dos filmes no momento, mas gosto muito de receber dicas de acordo com o meu gosto e conhecer novos gêneros e histórias.",
" Eu considero justas quando uma indicações possuem gêneros semelhantes,pois aí vejo uma lógica nas indicações,mas houve momentos que algumas indicações não fizeram sentido,porém como esse é uma espécie de estudo é totalmente compreensível que isso aconteça. Acho que o sistema pode sim ser uma boa alternativa de indicação",
" Se fossem filmes totalmente aleatórios e nem perto dos meus gostos pessoais",
" Muitos filmes que nunca nem ouvi falar",
" Foram justas",
" As recomendações dos filmes seriam injustas se nela aparecesse filmes nada parecidos com os da minha preferência",
" Eu acho que as recomendações seriam injustas no caso de você querer explorar filmes que você geralmente não assiste, pois com esse tipo de algoritmo você fica preso numa bolha, por exemplo eu já tinha visto mais da metade dos filmes que me foi recomendado, pois essa seria a minha bolha.",
" Concordo pouco, pois muitos dos filmes acabam sendo os mais vistos nos mesmo ano, e são poucas as variedades dos filmes escolhidos,para os filmes que são indicados.",
" As recomendações foram muito boas. Então acho que estavam muito bem nesse algoritmo.",
" Achei injustas, porque a escolha dos filmes foram escolhidos por algoritmos.",
" As recomendações são justas quando por exemplo, você avalia positivamente o filme, ou quando os assiste até o final na plataforma de streaming. Muita das vezes são recomendados filmes baseados em outros que foram abandonados, ou até mesmo pelo simples fato selecionar determinado filme, mas sem nunca chegar a assistir, apenas por ler a sinopse, por exemplo.",
"As escolhas dos filmes geradas pela Inteligência Artificial podem ser um pouco problemáticas devido ao fato de você se manter na sua bolha. Se uma pessoa que assiste só comédia acessar esse sistema, há uma grande chance do AI recomendar apenas comédias para ela.",
" Não acredito que tenha injustiça nessa escolha, se o algoritmo for programado de maneira correta, ele sempre irá me indicar filmes que são do meu agrado, mesmo q nunca tenha assistido. Só acharia injusto se o algoritmo mostrasse apenas filmes de determinado país, desconsiderando as outras culturas",
" A maior parte dos filmes que eu avaliei, foram os que me foram recomendados depois, então nas recomendações só teve 1 filme que eu não tinha avaliado e que eu não conhecia, mas fiquei com vontade de assistir por ser de um gênero que eu gosto. Então, eu achei as recomendações justas, só acho que o sistema poderia recomendar mais filmes, sem repetir tanto os que eu já avaliei.",
" A seleção foi fiel ao meu gosto. A maioria foram filmes que já assisti e gostei. Acabou sendo pouco útil, mas realmente foram compatíveis com meu gosto. Talvez recomendar filmes menos conhecidos seja um bom caminho.",
" Consideraria injustas as recomendações de filmes que claramente não condizem com meu gosto pessoal, mas que foram recomendados com o intuito promover certas produções.",
"Se ele me recomendasse filmes somente com orçamento alto, seria injusto. Se ele me recomendasse somente filmes que eu já assisti também. Justo seria ele me indicar filmes de gênero parecido com minhas avaliações.",
" Uma recomendação que não tenha a ver com algo que eu tenha selecionado ou de algum gênero cinematográfico que eu não goste",
" Eu considero justa, quando o sistema sabe que selecionei, assisti e avaliei positivamente o filme, e o mesmo me dê opções do mesmo estilo ou similar. Muitos sistemas são injustos pois não levam em consideração a avaliação, e ou somente selecionam opções de filmes assistidos sem serem avaliados ou filmes apenas selecionado para ver o trailer.",
" Eu gostei das recomendações e tanto são úteis e agradam meu gosto que a maioria dos filmes indicados eu já assisti e gostei",
" Acho justo recomendar filmes com alto e baixo orçamento, tendo em vista apenas a sua categoria e não sua popularidade e orçamento. Acharia injusto que recomendassem apenas filmes com alta popularidade e orçamento.  Já assisti bons filmes que não eram tão populares ou conhecidos, e filmes não tão bons com uma popularidade impressionante.",
" Quando não levam em consideração a classificação indicativa.",
" Seria injusta se recomendasse apenas filmes da mesma franquia (exemplo: escolhi um filme Marvel e ele me recomendar todos os filmes da marvel). É justo quando ele se baseia nos meus gostos por determinado genero, pelos atores e etc.",
"  Acredito que acabam sendo injusta quando insistem por recomendar sempre os mesmos filmes e gêneros, mesmo quando você acabe procurando por outras opções, mesmo não tendo assistido de fato, pois se eu procuro é porque há interesse em outros filmes e gêneros. No entanto no geral, grande parte das recomendações são justas quando assisto um tipo de gênero as recomendações acabam sendo muito certeiras e você acaba não perdendo tempo na barra de pesquisa ",
" Acho que recomendar só pelo gênero é meio injusto. Às vezes filmes do mesmo gênero são completamente diferentes.",
"  Eu as consideraria injustas se não fosse levado em consideração os gêneros dos quais havia selecionado com melhor classificação, em relação ao meu gosto. Considero escolhas justas, os casos que além de selecionar obras semelhantes aos gêneros que tenho mais intimidade, também me direcionar a obras que ainda não são tão populares.",
" Não gostei das sugestões",
" Justas.",
" Me foi recomendado filmes sequenciais aos que eu tinha marcado, achei justo visto que é o que gosto e o que teria interesse em assistir. Não seria acertivo me recomendar um gênero em que não mostro ao agoritimo interesse.",
" Seria injustas se selecionaram somente filmes que são muito assistidos, pq existem produções que não são tão conhecidas, porém são muito bons e podem se encaixar no gosto da pessoa.",
" o fato de alguns filmes serem pouco populares, o que desencadeia uma menor interação na recomendação deles",
" Algumas escolhas foram interessantes, mas outras eu não escolheria nem de longe e várias eu já tinha visto. Então concordo em alguns pontos e outros não.",
" Acho que são justas, já que levaram em conta o gênero escolhido e os filmes que avaliei. Seria injusto se seguissem um caminho que não considerasse as minhas escolhas.",
" Para mim não existe recomendação injusta, pois cada um decide se irá assistir ou não ao filme depois da recomendação.",
" Eu achei que o sistema recomendou filmes de acordo com meu gosto.",
" Algumas vezes não gosto das sugestões da Netflix",
" Fiz avaliação de velozes e furiosos 6 e me recomendaram Velozes e furiosos 5.",
" Na maioria das vezes, eu tento assistir aos filmes para ter uma opinião sinceramente sobre o mesmo e não acreditar totalmente na crítica. Porque sabemos que gosto difere muito de uma pessoa para outra.",
" Achei o sistema muito bom, apenas os filmes que estavam nas opções que para mim eram ruins.",
" Injustas em caso da não observação dos gêneros e características envolvidos, em casos de seletividade não justificada de orçamento/popularidade",
" Injusta quando ele desconsidera as demais escolhas. Eu escolhi umas duas animações e o restante foi filme e as recomendações foram praticamente só animação.",
" Talvez por ter escolhidos filmes antigos, o algoritmo me direcionou para esse tipo de filmes, escolhendo assim alguns títulos que não curti.",
" Ele só vê os filmes que vc deu maior nota e coloca a maioria dos filmes todos parecidos.... parece a seleção do Netflix....assisti um filme e eles dão sugestão de todos os filmes parecidos com aqueles que vc viu",
"Acredito que as indicações foram realizadas de forma justa exatamente por ser baseadas nas seleções do usuário.  Acredito que a única injustiça que pode ocorrer é referente a atualidade dos filmes indicados, visto que já assisti a grande maioria. No entanto, isso pode ser algo para o meu caso em específico.",
" Quando não levassem em conta as minhas preferências",
" Nesse caso. Mais da metade dos filmes eu nunca havia assistido, tive que pesquisar.",
" Se assisti um único filme de determinado gênero não significa que os próximos vou querer aquele.. nesse caso as recomendações seriam injustas... mas não tem como o programa identificar esse tipo de coisa",
" Eu gostei das sugestões e não tenho nenhuma observação para fazer",
"  Os algoritmos são utilizados de forma ""injusta"" quando estão de forma excessiva, onde acredito que influencie bastante no consumismo, que muitas vezes não seja necessário. Mas ele se torna justo nos casos onde o indivíduo se interessa em algo e gostaria de ter mais informações e sugestões, como no caso deste estudo.",
" Não achei nenhuma recomendação injusta, só teria considerado injusto se eu não tivesse assistido ou não poderia assistir nenhuma recomendação.",
"  Injustas quando o filme não estar entre os gêneros que a pessoa gosta. Justas quando o filme é do gênero que a pessoa gostou ou parecido.",
" Claro que existem inúmeras variáveis no momento de escolha de um filme. No geral, o sistema é muito bom.",
" Achei as recomendações justas, apesar de preferir gênero romance e ter tido poucas indicações deste.",
" Acredito que a seleção ocorreu naturalmente de acordo com os. Filmes que eu havia selecionado e de acordo com meu gosto pessoal. No final, as indicações tinham a ver com o tipo de filme que eu consumo então não considero injusto.",
" Quando alguém avalia apenas um tipo de filme o algoritmo acaba indicando apenas o mesmo tipo, sendo injusto com outros temas.",
" Justas : pois levam em consideração as suas escolhas previas  Injustas : se fossem recomendar filme que não tem nada haver só pra fazer ele ter mais audiência",
" No caso de as recomendações serem sempre repetidas, limitando as possibilidades do usuário",
" Fazendo pergunta específica sobre o perfil sesejado",
" Achei a seleção justa visualmente falando, porém fiz 10 escolhas na fase inicial onde a maioria se tratava de animações e comédias românticas, creio que esse tipo de escolha abre uma ampla gama de indicações, apesar de conhecer quase todos os filmes da indicação final a maioria não se tratava dos gêneros que escolhi, mas de filmes que tiveram grande repercussão para o público em geral.",
" Acredito que o caso de seleções injustas possa ocorrer quando a pessoa respondendo o questionário tem um gosto muito abrangente para filmes, possivelmente fazendo com que apareçam indicações que não entrem em seu gosto pessoal. Já as justas, acredito que ocorram quando as escolhas da pessoa estejam em estilos próximos. No meu caso, as indicações foram principalmente de animações, porém dentro delas havia diferentes tipos de histórias, desde aventura à comédia.",
"  Eu acredito que independente do sistema no final a decisão é do consumidor, o algoritmo irá ajudar a sair da zona de conforto ao recomendar um filme, sugerindo algo que você ainda não teve a oportunidade de assistir, mas quem decide no final se irá ou não seguir essa recomendação é o consumidor. Deste modo acredito que essas recomendações sejam justas ao facilitar a decisão final do consumidor.",
" Quando eles não fossem semelhantes, levando em consideração principalmente o roteiro",
" Talvez em um momento em que não tenhamos nada para fazer ou assistir.",
" Não considerei injusta, todos os filmes exceto um eu já havia assistido e gosto muito de todos",
" Acredito que os aplicativos recebe mendam mais de acordo com o gênero e não com base na história o q pode ser um pouco injusto",
"  Me deu boas sugestões, porém 80% já havia assistido, sem uma opção de atualizar a página novamente para novas sugestões, porém gosto de todos que foram sugeridos e assisti. Acho justo as recomendações dos algoritmos, afinal é o que quero assistir naquele momento, é o meu ""padrão"". Porém acho injusto que não tenha uma opção de Reset para poder sair da zona de conforto e padrão de filmes que tenho assistido.",
" Justa",
" Acho que essas recomendações foram muito justas, pois foram recomendados filmes considerados pouco conhecidos e com um orçamento não tão alto, mas as vezes, acredito que essa sensação de injustiça possa ocorrer pois apenas filmes mais conhecidos e bem avaliados quase sempre são os sugeridos, quando procuramos por recomendações.",
" Achei desatualizados, poderia haver tanto filmes mais antigos quando atuais.",
" Muitos filmes que já assisti e que já havia avaliado.",
" Desde que as recomendações sejam baseadas somente em meus gostos e opiniões pessoais, são boas recomendações. Seria injusto se aparecece somente recomendações pagas.",
" Nos momentos em que você não tem nenhuma noção do que quer assistir.",
" Injusta se recomendassem filmes totalmente diferentes dos quais selecionei no início.",
" Filmes baseados nas minhas escolhas",
" Injustas quando há filmes que vc com toda certeza não assistiria.",
" As opções na primeira respostas eram limitadas, assisti poucos dos filmes recomendados, podendo induzir ao erro o programa, me levando a ver filmes dos quais não gostaria",
" Todos os filmes recomendados foram filmes que já assisti e que gosto, acredito que as recomendações foram justas e fiéis às minhas preferências",
" Seria injusto se recomendasse apenas um gênero",
" Seriam injustas se mesmo com a pesquisa o algoritmo indicasse filmes que não tem a ver com o meu gosto.",
" Quando eu gosto de um gênero específico e o algoritmo me recomenda o completo oposto",
" Acredito que poderia ter mais tipos de filmes paraamalisar e criar uma seleção de sugestões mais concreta. Talvez, a exclusão de gêneros que não assiste.",
" Seria injusta se fosse uma recomendação que não tem relação com minhas preferências.",
" Sendo baseado num avaliação prévia de minha escolhas, não acredito que srja injusto, mas deve ser uma escolha do usuário optar por este sistema quamdo desejar. Apesar recomendar filmes de minha total preferencia, eu não assisto filmes unicamente de um mesmo tema, das recomendações feitas a maioria dos filme foi de animação, apesar de adorar não é o tema que vejo exclusivamente. Ao recomendar com base em minha preferencia, exclui outra possibilidades que teriam me agradado mais.",
" Eu não considerei elas injustas, mas não tenho certeza se veria todos aqueles filmes",
" Depende , gosto de outros gêneros tb .",
" Eu achei as recomendações justas pois me recomendou filmes tanto de baixo quando de alto orçamento e filmes parecidos com os que avaliei anteriormente. Outro ponto é que ele reconheceu os meus gostos e, apesar de muitos dos filmes que me recomendou eu já tinha assistido, 2 filmes recomendados que eu ainda não tinha assistido me interessou e pretendo assistir.",
" Pra mim foi recomendado filmes que realmente eu gostaria de assistir, o que eu selecionei por exemplo foi um filme que eu já assisti e amo de coração, que é “estrelas na terra” eu gostei bastante do App",
" Eu achei o sistema confuso e não vi como uma forma de agregar com sugestões que combinassem com o meu gosto em geral",
" Coração de tinta apesar de ser pouco conhecido é um dos meus favoritos e é interessante como a IA conseguiu chegar nele",
" Só podem ser consideradas injustas se na seleção inicial não tiverem muitas opções para validar os gostos pessoais.",
" Acho que as recomendações são justas, algo que facilita na escolha",
" Não foram muito condizentes ao meu gosto atual.",
" Seria injusto as recomendações caso o sistema pertencesse a um dos grandes estúdios de cinema e só indicasse filmes desse estúdio. Nas recomendações que recebi, havia filmes de estúdios diferentes e todos estavam dentro do meu gosto.",
" Na maior parte do tempo filmes já comumente conhecidos vão ser recomendados com mais faculdade que outros.",
" Seria injustas quando não são baseados nas decisões feitas e são justas quando se baseiam.",
" Injustas: filmes sem nenhuma conexão com os escolhidos anteriormente. Justa: filmes com temáticas parecidas aos que já assisti.",
" Justa, as escolhas foram de acordo com o meu gosto, só notei que foi adicionado filmes de gostos gerais na lista, mas não foi um problema.",
" Apesar de ter assistido quase todos filmes indicados, não acho que as recomendações foram justas ou injustas, pois os filmes indicados estão dentro do meu estilo. Minha única observação é que os filmes da primeira tela não são muito conhecidos e as vezes difícil de avaliar por ela, pois a maioria eu não conhecia.",
" Nao acho que recomendações sejam injustas, vai depender do gosto de cada um",
" Quando as recomendações indicam filmes que eu dei nota anteriormente, acaba por se tornar ineficiente,uma vez que eu já assisti os referidos filmes.",
" Injustas se me recomendasse algo que tenha algum ator ou diretor que eu não gosto muito",
" Pra mim foi justo sugerir filmes pouco conhecidos e de baixo investimento, as vezes são ótimos e nós jamais o descobririamos de outra forma. Injusto seria coloca-los lado a lado com grandes obras que não deixariam chance à essas pequenas artes.",
" As vezes a recomendação baseada nas suas preferências é boa ,mas mesmo assim você pode não gostar muito do filme, acontece....",
" Justas quando traça o perfil e se baseia nele , injusto quando sugere aleatoriamente",
" Não entendi a pergunta",
" Acredito que seja injusta quando só recomenda filmes mais conhecidos e com alto orçamento e mais justo quando inclui filmes menos conhecidos tbm.",
" Eu nunca usei algoritmos para recomendação de filmes. Outras demandas de recomendação eu gosto de usar não tenho do que reclamar. As recomendações que tive aqui foram boas pela descrição dos filmes que não conheço e vou assistir. ☺️",
" Eu não gostei das sugestões de comédia porque não gosto muito desse tipo de filme, mas consigo entender porque foram sugeridos, com base nas minhas respostaa",
" Em casos que o algoritmo sugerisse títulos com enredo totalmente diferentes do que os que eu escolhi",
" Injustas se só tivesse filmes mais populares, justa pois apareceu filmes que nunca assisti porém achei boa as recomendações",
" Não achei nenhum caso em que a recomendação foi injusta",
" Acho as recomendações injustas quando eu apenas clico em algum filme para ler a sinopse ou ver o trailer mas o algoritmo passa a me recomendar o mesmo tipo ou gênero. Deveria faze-lo apenas quando eu desse play no filme.",
" Normalmente não tenho problemas com o algoritmo. Visto que vejo filmes de terror, séries coreanas românticas, filmes românticos, e o algoritmo sempre me ajudou nas buscar por filmes recomendados para mim e que realmente eu tenho uma alta probabilidade de assistir.",
" Achei um pouco difícil a primeira etapa ,pois demorava muito pra altera o tema : romance,drama",
" As escolhas apresentadas não seguem minhas preferências, não é signifiticamente fiel",
" Algumas recomendações foi de filmes avaliados anteriormente.",
" Tem bastante opção de filmes acharia injusto se não tivsee bastante filmes",
" Injusta : gêneros que não foi selecionado e que fogem totalmente da premissa dos que assisti. Justas: os gêneros e as temáticas dentro das opções",
" Acho que as recomendações, foram de acordo com a classificação que foram atribuídas as filmes dos primeiros questionário, e apareceram boas recomendações de acordo com o que eu pontuai que era o que mais gostava, a opção mostrou filmes que eu já assisti e estão de acordo com a classificação. De forma geral, eu gostei da plataforma e provavelmente usaria ela",
" Injusto pois a maioria dos filmes são hollywoodianos, falta apresentações de filmes mais alternativos",
" Consideraria mas justas se as recomendações fossem de filmes não tão populares",
" Não tenho opinião sobre isso.mas gostei do sistema",
" Acho injusto quando limita as opções baseadas em apenas um gosto, visto que no começo foram poucas as opções de que já tinha assistido, então se basear apenas neles foi um pouco incoerente, pois meu gosto abrange mais que aquelas opções.",
" Consideraria injusta a recomendação de um filme que nunca demonstrei nenhum tipo de interesse, seja pelo gênero, atores, diretor",
" Quando os filmes não tem nada a ver com o que a pessoa escreve como recomendação, aí seria injusta.",
" Considerei neste sistema, muito justa. Acertaram totalmente as minhas opções que eu assistiria.",
" Justas",
" Essas recomendações são justas pois são baseadas em nossas escolhas e gostos pessoais mas pode ser injusta com base em algo genérico e popular",
" Não gosto muito de filme de açaí, o quão foram me recomendados",
" Quando elas são baseadas em orçamentos",
" Nos casos dos filmes de terror",
" Todos já vi, então se tornou irrelevante as sugestões.",
" Acho que o que apareceu está dentro do aceitável, de acordo com o que eu havia escolhido inicialmente.",
" Com foco no meu gosto pessoal, as recomendações foram justas",
" Não sei dizer ao certo foi a primeira vez que utilizei essa tecnologia",
" quando não fazem questão de saber o gosto da pessoa q vai assistir",
" Recomendações baseadas nas sequências dos filmes avaliados previamente pelos usuários são injustas, visto que elas estão subentendidas se o usuário já marcou que assistiu ao primeiro filme. Recomendações mais justas devem se basear em similaridades de enredo e gênero.",
" Não saberia opinar",
" Tem filme que achei que fosse antigo",
" As recomendações não foram muito boas porque os filmes que tive que avaliar eram muito limitados. Eu só sonhecia 2, então tive que falsear 5 avaliações extras. Se melhorarem os filmes que temos que avaliar realmente, creio que as recomendações da IA fiquem bem melhores e batam com os gostos da pessoa.",
" No caso de ter poucas opções dos filmes que você realmente assistiu, tive essa dificuldade em achar os filmes",
" Casos de injustiça seria por exemplo: no momento da avalição, eu avaliar somente filmes de animação e como chegar na recomendação aparecer filmes de terror por exemplo",
" Eu considerei bem justa na seleção do filme as branquelas e outro que era esposa de mentirinha amo esses filmes ,então na minha opinião foi uma seleção justa .",
" Justas: quando queremos compartilhar esses dados. Injustas: quando não queremos compartilhar dados.",
" Eu consideraria injustas se houvessem privilégios ou restrições já estabelecidas pelo sistema",
" Achei que foram opções muito justas para meus gostos. Ele seguiu certinha o que eu gosto num filme, e salvei o nome de alguns para ver depois. Acho que podem ser injustas caso o usuário não goste de um gênero específico, mas gostou de um filme deste gênero. Não seria culpa da I.A. e sim do usuário.",
" .",
" as recomendações poderiam ser mais váriadaspara dar a opção de ver outros filmes",
" Filmes mais popular tem mais chances de ser indicados",
" Gostaria de mais opções de buscas de filmes",
" Quando é gerado filmes que não se adequam ao perfil daquele que responde",
" Recomendar um filme continuação somente porque a pessoa assistiu o primeiro não seria justo.",
" Eu considero justas quando o filme recomendado é baseado no seu gosto, e injusto quando o filme recomendado, não tem nada haver com seu gosto",
" Não consigo ver dessa maneira. Se está me ajudando a dar opções pela qual costumo seguir, se não gostar da recomendação de algum vou para outro recomendado que goste.",
" Acredito que tenha que ter uma variedade maior de opções com todos os estilos de filmes, para que assim as indicações seja mais próximo do que a pessoa goste.",
" Em questões de recomendações de filmes, séries e livros gosto muito. Apenas não gosto quando aparece muita propaganda.",
" Injusta no caso de não levar meus gostos em consideração.",
" No caso o algoritmo escolheu a sequência dos filmes que avaliei e filmes baseados em livros, o que foi justo e conforme minha preferência. Para diversificar mais as escolhas poderiam ter opções parecidas com os filmes avaliados e não apenas a sequência deles.",
" Eu pensaria que era injusto se me recomendasse um terror, já que não é da minha preferência, mas desenho acertou em cheio, na verdade acertou todos, foram filmes que eu já vi e vejo novamente de tão bons que são. Eu amei as recomendações.",
" Consideraria injusta se me recomendasse filmes de ação. Não gosto. Prefeitos romances leves ou bem clichê, tais como filmes q exaltam a feminilidade",
" Justas",
" Recomendações que fossem muito divergentes do meu gosto, ou com avaliações ruins.",
" Havia filmes ali que não me interessaria assistir, já outros que com certeza assistiria ...",
" Acredito ter sido justa, pois me foi indicado muitos filmes que já fazem parte do meu grupo de filmes assistidos ou que desejo assistir.",
" Acho que recomendar filmes fora da faixa etaria do usuario",
" Recomendou filmes uma poucos antigos, os de animações nao importa o tempos, mas os de ação e comedias românticas pra mim foram um desconhecido por causa do tempo em que elas foram lançadas, esse sistema poderia separar por data de lançamento, tipo uma locadora",
" Acredito que no banco de dados não estavam cadastrados meus filmes preferidos",
" Injusta se não tivesse nenhum que eu gostasse.",
" Justas porque me recomendaram dentro do que tinha escolhido antes: animações e musicais. O que posso dizer de injusto seria somente imprecisão, pois sugeriu coisas diferentes do que mostrei gostar",
" Não sei",
" Baseado nas minhas escolhas e pesquisas entendo que as recomendações condizente com as mesmas com algumas excessões que não dão match total com minha escolha entre as seleções. Assistiria total mas entre outras opções ele não seria a escolha.",
" Ao procurar por um determinado filme, de um determinado gênero, se espera que as sugestões sejam de gêneros meramente iguais. Se eu procuro por Harry Potter, por exemplo, quero que as sugestões baseadas sejam de aventura, adolescência, magia, sobrenatural. Isso é considerado justo.",
" Filmes de gêneros que nunca assisti ou não assistiria, filmes que não tem nada haver com os gêneros que assisto",
" Achei justas, porém as recomendações de romances como são a maioria do gênero que assisto, eu já tinha assistido. E continuo tendo maiores interesses nesse gênero, aí não apareceu muita coisa nova.",
" Considero justo quando assisto um filme e são recomendados filmes parecidos ou do mesmo estúdio, como Ghibli por exemplo. Injusto quando são recomendações totalmente diferentes do que eu assisti.",
" Acho que seriam injustas caso utilizasse meus dados de alguma forma para obter um bom resultado. Acho que o fato de fazerem uso apenas das minhas respostas para chegar num resultado torna a IA justa.",
" Pra mim não se viu pois dos 6 primeiros filmes eu só tinha visto um. Então ele não conseguiu comparar e me indicar bem os filmes que eu veria.",
    " Primeiro, nas opções de filmes assistidos eu selecionei alguns de animação mas tentei variar entre comedias e alguns filmes do mesmo gênero que amo e odeio (como Tróia a 300), porém eu observei que a maioria das sugestões de filmes eram animações, somente um filme era de outro gênero! Além de serem massivamente animações todas eram de estúdios grandes (Disney, Pixar e DreamWorks, etc), assim achei injusto pois ao meu ver (de pessoa leiga no assunto kkkk), achei que o algoritmo favoreceu as obras de grandes estúdios e mais populares, particularmente não achei que expandiu minhas recomendações, no entanto, achei extremamente certeira as recomendações, tanto que ja vi a maioria, e senti genuíno interesse pelo único filme recomendado que não tinha visto! No geral é isso, boa sorte <3",
" Acho que ainda não sei opinar ao certo sobre o assunto, acredito que as opções foram interessantes e baseadas nas minhas preferências certinho.",
" A maioria foi justa",
" Pode ser injusta quando não sabe exatamente os gostos da pessoa e pode ser justo quando seleciona ótimos filmes de qualidade que a pessoa gosta a ponto de indicar para mais pessoas",
" Achei justa por mais que a maioria tivesse sido continuações de filmes que selecionei.",
" Acabei avaliando mais filmes de animação e romance, e o algoritmo acabou recomendando animação e eu já havia assistido a todos, e dentre as opções haviam duas comédias coranticas nas quais eu tbm já tinha assistido.  Achei justa as recordações de filmes, mas achei injusto que as recordações acabaram aparecendo apenas opções nas quais eu já tinha assistido. Mas é uma forma diferente de se escolher um filme quando se está na dúvida.",
" Não sei responder",
" Não tenho uma opinião formada sobre.",
" Não considerei né justa e né injusta, pode ser que futuramente eu assista ou não.",
"  Injustas: quando mostram apenas filmes americanos. Não indicando filmes de outras nacionalidades.  Justo: quando mostram filmes que realmente estão de acordo com meu gosto.",
" Acredito que um algoritmo de recomendação de filmes não faça recomendação injustas, pois se baseia em algo que você já assistiu. Todovia, a análise devia ser mais minuciosa para eliminar os que não curtimos muito, tornando mais justa a recomendação, e indicar diretamente o próximo da sequência não todos da franquia, para mostrar outras opção caso não deseje continuar naquele universo.",
" A maioria dos filmes eu assistiria , então considero justas",
" As recomendações injustas seriam as que consistiam nos mesmos filmes que eu avaliei, pois são recomendações de filmes que eu obviamente já assisti, embora goste muito deles não deixa de ser uma recomendação repetida. As recomendações justas são as de filmes de estilos parecidos com os que eu avaliei.",
" Se fosse escolhido filmes totalmente diferentes do estilo de filme que avaliei",
" Injustas se fossem restritas a blockbusters. Justas se forem pra mostrar algo que usualmente não encontraríamos.",
" Quando o filme é pouco popular e não é recomendado",
" Segundo a critica",
" Eu entenderia como uma recomendação injusta se me indicasse um tipo de filme que eu não assistiria, como de terror por exemplo. Considerando que nenhuma opção como referência apontou que eu tinha interesse nesse tipo de gênero.",
" Acredito que para ser uma recomendação injusta, as recomendações iriam ser de gêneros que eu não gosto.",
" Considero as recomendações de filmes justas quando a pessoa escolhe utilizá-las conscientemente, ou seja, as recomendações não aparecem de forma não solicitada e o indivíduo sabe que as recomendações vêm de uma inteligência artificial.",
" Justas quando faz jus aquilo que eu gosto",
" Seria injusto se não levasse em consideração a opinião, se recomendasse um filme que teve uma péssima votação.",
" Todas as recomendações foram justas",
" Se fosse algo extremamente fora de minhas preferências",
" É um pouco injusto na questão de que dificilmente as pessoas sairiam da sua bolha de “apreciação”, visto que o algoritmo sempre recomenda aqueles que estão no rol do gosto popular",
" Justas. Porém previsível",
" Acho injustas quando as recomendações não selecionam filmes poucos conhecidos e só indicam os populares",
" Para falar a verdade eu escolhi filmes com continuação, ou filmes como Batman por exemplo, que já foi lançado em várias versões então o sistema entendeu isso, porém eu gostaria que recomendasse filmes do mesmo gênero, porém diferentes dos que eu recomendei.",
" Quando escolhesse gênero totalmente oposto ao gosto do que o indivíduo consome. Quando os enredos fossem muitos discrepantes do que a pessoa gosta de assistir",
" Seria injusto se os filmes recomendados forem totalmente diferentes dos gostos que foram informados na recomendação. (ex : escolho comédia, e os filmes recomendados são terro) enquanto justo é basicamente recomendar um filme com base nos gostos apresentados. Espero que tenha sido útil.",
" Na minha humilde opinião, todos os filmes que foram recomendados foram vistos e revisto milhares de vezes. Eu achei bastante justa essas recomendações.",
" Não achei nada injusta, pelo contrário, as indicações foram justar e inteligentes.",

]



def codificar_respostas_melhoradas(respostas):
    codificacoes = []

    for resposta in respostas:
        codificacao = {"Tema": []}



        # coding
        if "justa" in resposta.lower() or "hypados" in resposta.lower() or "conhecidos" in resposta.lower():
            codificacao["Tema"].append("Justica")
        if "preferências" in resposta or "perfil" in resposta or "acordo" in resposta or resposta.find("gosto") != -1 or resposta.find("parecido") != -1:
            codificacao["Tema"].append("Acuracia")
        if  "diferentes" in resposta.lower() or "temática" in resposta.lower() or "diversidade" in resposta.lower() :
            codificacao["Tema"].append("Diversidade")
        if "orçamento" in resposta.lower():
            codificacao["Tema"].append("Diversidade de Filmes")
        if resposta.find("pouco") != -1 and resposta.find("conhecido") != -1:
            codificacao["Tema"].append("Inclusão de Filmes Menos Conhecidos")

        codificacoes.append(codificacao)

    return codificacoes



```
